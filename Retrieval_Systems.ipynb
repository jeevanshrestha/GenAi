## Vector Space Model

```python
vectorizer = TfidfVectorizer()
```

```python
tfidf_matrix = vectorizer.fit_transform(preprocessed_docs)
```

```python
print(f"the shape of tfidf_matrix is {tfidf_matrix.shape}")
```

```python
query = "the door"
query_vec = vectorizer.transform([query])
similarities = cosine_similarity(tfidf_matrix, query_vec).flatten()
```

```python
similarities
sorted_similarities = list(enumerate(similarities))
sorted(sorted_similarities, key=lambda x:x[1], reverse=True)
```

```python
def search_tfdif(query, vectorizer, tfidf_matrix):
  #vectorizing th query
  query_vec = vectorizer.transform([query])

  #computer the cosine similarty
  similarities = cosine_similarity(tfidf_matrix,query_vec).flatten()

  #pair each document index wit similarity score
  sort_similarities = list(enumerate(similarities))

  #sort the documents by similarity score
  results = sorted(sort_similarities, key=lambda x:x[1], reverse=True)

  return results
```

```python
#Apply the function to the query
query = "chamber least"
search_results = search_tfdif(query, vectorizer, tfidf_matrix)
search_results[:10]
```

```python
#Print out the documents
print(f"Top 10 documents by similarty score for query '{query}'")
for doc_index, score in search_results[:10]:
  print(f'Document {doc_index + 1}: {docs[doc_index]}')
```

```python
#import libraries
import os
import shutil
from whoosh.index import create_in
from whoosh.fields import *
from whoosh.qparser import QueryParser
```

```python
def preprocess2(text):
  tokens = nltk.word_tokenize(text)
  tokens = [word.lower() for word in tokens if word.isalnum()]
  #Define the stop words
  stopwords =set(nltk.corpus.stopwords.words('english')) - {"and", "or", "not"}

  return [word for word in tokens if word not in stopwords]
```

```python
 text ="hello world with the about any of these are't because this is very important in the sense of all"
 preprocess2(text)
```

```python
#create a new folder , remove the old one if available
if os.path.exists("index_dir"):
  shutil.rmtree("index_dir")
os.mkdir("index_dir")
```

```python
#Define a schema for the directory

schema = Schema(title = ID(stored=True, unique=True),
          content = TEXT(stored=True))
```

```python
#Create index in the folder
index = create_in("index_dir", schema)
```

```python
writer = index.writer()
for i, doc in enumerate(docs):
  writer.add_document(title=str(i), content=preprocess2(doc))
writer.commit()
```

```python
# Query using boolean search

query = "chamber least"
parser = QueryParser("content", schema=index.schema)
parsed_query = parser.parse(query)
print(f"Parsed query: {parsed_query}")
```

```python
with index.searcher() as searcher:
  results = searcher.search(parsed_query)
  print([[hit["title"], hit["content"]] for hit in results])
```

```python
def boolean_search(query, index):
  parser = QueryParser("content", schema=index.schema)
  parsed_query = parser.parse(query)
  with index.searcher() as searcher:
    results = searcher.search(parsed_query)
    return [(hit['title'], hit['content'])  for hit in results ]
```

```python
results = boolean_search(query, index)
results
```

```python
#Tokenize the documents

tokenized_docs = [preprocess(doc) for doc in docs]
tokenized_docs
```

```python
#Initialize the BM25 MOdel
bm25 = BM25Okapi(tokenized_docs)
```

```python
def search_bm25(query, bm25):
  tokenized_query = preprocess(query)
  results = bm25.get_scores(tokenized_query)
  return results
```

```python
import numpy as np
query = "chamber least"
results = search_bm25(query,bm25)
sorted_results = np.argsort( results )[::-1]
```
